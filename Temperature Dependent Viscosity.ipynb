```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = [8, 5]
plt.rcParams["figure.autolayout"] = True
plt.gca().invert_yaxis()
y = np.linspace(1, 0,10000)
y1 = np.linspace(1, 0,10000)
y2 = np.linspace(1, 0,10000)
y3 = np.linspace(1, 0,10000)
upper_bound = 6.5051499783199
x1 = np.log10(np.exp(6.6*((1/(y+0.11))-(1/(1.11)))))
x2 = np.log10(np.exp(8.8*((1/(y+0.11))-(1/(1.11)))))
x3 = np.log10(np.exp(9.9*((1/(y+0.11))-(1/(1.11)))))
x1 = x1[x1 <= upper_bound]
x2 = x2[x2 <= upper_bound]
x3 = x3[x3 <= upper_bound]
y1 = y1[:len(x1)]
y2 = y2[:len(x2)]
y3 = y3[:len(x3)]
#print(y1[-1], y2[-1], y3[-1])
plt.plot(x1, y1, color='red', label = "A = 6.6")
plt.plot(x2, y2, color='blue', label = "A = 8.8")
plt.plot(x3, y3, color='green', label = "A = 9.9")
plt.axvline(x=6.5051499783199, color = 'black', label = "$\eta_{cutoff}$ = 3.2e6", ls = '--')  #6.5051499783199 is log(3.210^6)
plt.vlines(x=6.5051499783199, ymin = 0, ymax = 0.20542054205420535, colors = 'red', alpha=1) #line for y1
plt.vlines(x=6.5051499783199, ymin = 0, ymax = 0.2742274227422742, colors = 'blue', alpha=0.80) #line for y2
plt.vlines(x=6.5051499783199, ymin = 0, ymax = 0.3043304330433043, colors = 'green', alpha=0.70) #line for y3
plt.legend()
plt.xlim(xmin=0 ,xmax=7)
plt.ylim(ymin=1 ,ymax=0)
plt.xlabel("Temperature dependent viscosity, $log(\u03B7_T)$", fontsize=16)
plt.ylabel("Non-dimensinal Temperature", fontsize=16)
plt.title("Temperature dependent viscosity for different non-dimensional activation energy", fontsize=13)
#plt.savefig("Different non-dimensional activation energy with Eta cutoff.png")
```
